## ë°ì´í„°ì…‹ ê²½ë¡œ ì„¤ì •

```python
from glob import glob

# ì´ë¯¸ì§€ ê²½ë¡œ listë¡œ ë„£ê¸°
train_img_list = glob(dataset.location+'/train/images/*.jpg') + glob(dataset.location+'/train/images/*.jpeg')
valid_img_list = glob(dataset.location+'/valid/images/*.jpg') + glob(dataset.location+'/valid/images/*.jpeg')
test_img_list = glob(dataset.location+'/test/images/*.jpg') + glob(dataset.location+'/test/images/*.jpeg')


# txt íŒŒì¼ì— write
with open(dataset.location+'/train.txt', 'w') as f:
	f.write('\n'.join(train_img_list) + '\n')

with open(dataset.location+'/valid.txt', 'w') as f:
	f.write('\n'.join(valid_img_list) + '\n')

with open(dataset.location+'/test.txt', 'w') as f:
	f.write('\n'.join(test_img_list) + '\n')
```

```python
import yaml
with open(dataset.location+'/data.yaml', 'r') as f:
  data = yaml.load(f, Loader=yaml.FullLoader)
print(data)

data['train'] = dataset.location+'/train.txt'
data['val'] = dataset.location+'/valid.txt'
data['test'] = dataset.location+'/test.txt'

with open(dataset.location+'/data.yaml', 'w') as f:
  yaml.dump(data, f)
print(data)
```
